```python
actions = ['obey', 'obey_strange', 'ignore', 'hide_or_raise', 'ignore_and_google']
backgrounds = ['gender', 'blood', 'age']

for i in backgrounds:
    for j in actions:
        plt.figure()
        sns.barplot(dd[i], dd[j])
```

```python
for i in backgrounds:
    for j in actions:
        plt.figure()
        sns.catplot(x=i, y=j, data=dd)
```
